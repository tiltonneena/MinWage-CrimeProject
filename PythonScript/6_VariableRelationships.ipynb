### 2. Exploring relationships - correlations

```python
df.corr()
```

Not all columns are needed for the correlation heatmap, make subsets with only neccessary columns:

```python
df.columns
```

```python
df_sub = df[['Year','state_mw', 'fed_mw', 'effective_mw','prisoner_count',
       'state_population', 'violent_crime', 'murder', 'robbery', 'burglary',
       'incarceration_rate', 'avg_rate_of_year']]
```

```python
df_sub.columns
```

Create correlation heatmap: 

```python
# Create a subplot using matplotlib
f,ax = plt.subplots(figsize = (10,10))

# Create the correlation heatmap for states subset using seaborn by applying heatmap onto correlation matrix and the subplots defined above. 
corrmap = sns.heatmap(df_sub.corr(), annot = True, ax = ax) 
```

Create similar heatmap but with divergent colors for different type of viewing: 

```python
# Create correlation matrix with subset of dataframe

sub_corr = df_sub.corr()
```

```python
# Make seaborn heatmap for subset
ax = sns.heatmap(sub_corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(150,275,s=80,l=55,n=9), square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
```

Notable relationships: Year & Avg Rate of (incarceration) per Year, Federal Minimum Wage & Avg Rate of (incarceration) per Year, State Minimum Wage & Incarceration rate.

### 3. Exploring relationships - scatterplots

Explore one of the relationships mentioned above using a scatterplot. 

```python
# Create a scatterplot showing relationship between 'fed_mw' & 'avg_rate_of_year'

sns.lmplot(x = 'fed_mw', y = 'avg_rate_of_year', data = df_sub)
```

The scatterplot is sparse, but there is a negative correlation between the two variable. However, there are quite a bit of data points far from the slope. 

### 4. Exploring relationships - pair plots

### 5. Exploring relationships - categorical plots

The dataframe already has assigned categories ranking incarceration rate (or percentage). The categorical plot will show the relationship between 'year' and 'incarceration rate'.

```python
sns.set(style = 'ticks')
g = sns.catplot(x = 'Year', y = 'incarceration_rate', hue = 'rate_rank', data = df)
```

```python
sns.set(style = 'ticks')
g = sns.catplot(x = 'Year', y = 'state_mw', hue = 'rate_rank', data = df)
```

There is no pattern indicating here that lower minimum wage correlates to higher incarceration rank. 

```python
sns.set(style = 'ticks')
g = sns.catplot(x = 'Year', y = 'burglary', hue = 'rate_rank', data = df)
```
